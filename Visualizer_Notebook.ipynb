```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_colors():
    return [0, 10, 60, 90]

def show_im(im, ax=None):
    colors = get_colors()
    temp_im = np.array(im, dtype=np.uint8)
    for i in range(len(colors)):
        temp_im[temp_im == i] = colors[i]
    if ax == None:
        plt.imshow(temp_im, cmap='terrain', vmin=0, vmax=255)
    else:
        ax.imshow(temp_im, cmap='terrain', vmin=0, vmax=255)
```

```python
tilemap = np.ones((11,10))

tilemap[0:2, :] = 3
tilemap[2:3, :7] = 3
tilemap[3:5, :5] = 3
tilemap[4, 5:7] = 3
tilemap[5, :3] = 3
tilemap[2, -1] = 3

tilemap[1:3, 1:3] = 2
tilemap[1, 3] = 2
tilemap[2,4] = 2

tilemap[7, 4:8] = 0
tilemap[8:10, 5:8] = 0

show_im(tilemap);
```

```python
tilemap = np.ones((5,5)) * 3

tilemap[2:5, 2:5] = 2
tilemap[1, 2] = 2
tilemap[2, 1] = 2

show_im(tilemap);
```
